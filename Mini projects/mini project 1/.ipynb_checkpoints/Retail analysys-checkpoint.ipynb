# Preprocessing

```python
df.isnull().sum()
```

```python
# Adding a new category of Other in the the Description column
df['Description'] = df['Description'].fillna("Others")

# Adding a new column of the Registration of the customer
df["Registered"] = df["CustomerID"].copy()
df["Registered"] = df["Registered"].fillna("Not Registered")
df["Registered"] = df["Registered"].apply(lambda x: "Registered" if type(x) == float else x)

# Fill the null values with zeros in Customer Ids
df["CustomerID"] = df["CustomerID"].fillna(0)

# Adding new column for the total price
df["Total_price"] = df["UnitPrice"] * df["Quantity"]

# Removing the records containing the negative values records
df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]

# Adding three new columns of year month and day
df["Invoice_Year"] = pd.to_datetime(df["InvoiceDate"]).dt.year
df["Invoice_month"] = pd.to_datetime(df["InvoiceDate"]).dt.month_name()
df["Invoice_day"] = pd.to_datetime(df["InvoiceDate"]).dt.day
df["hour"] = pd.to_datetime(df.InvoiceDate).dt.hour

df.head()
```

# Hypothesis

My insights sayas the the best selling in the whole world is from 9 am to 4 o clock

```python
sns.lmplot(x='hour', y ="Total_price", data=df);
```

we are going to consider sales between 6:00 to 20:00

Set alpha (Let alpha = 0.05)


```python
alpha= 0.05
```

let's create two sets of data

High Sales : If time is from 8 to 17
low Sales:  If time is from 18 to 7  
Take a sample of 100 record in each dataset.

```python
# ANSWER
N = 100
part1 = df[(df['hour'] > 8) & (df['hour'] < 17) ]['Total_price'].sample(N)
part2 = df[(df['hour'] > 1) & (df['hour'] < 8) | (df['hour'] > 17)]['Total_price'].sample(N)
```

```python
ax = sns.distplot(part1, label = 'High Sales', kde= False);
sns.distplot(part2, ax = ax, label = 'Low Sales', kde = False);
plt.legend(loc = 'upper right')
plt.xlabel('')
plt.ylabel('')
plt.show();
```

```python
from scipy import stats
stats.ttest_ind(part1, part2)
```

**Variance**

```python
# ANSWER
print("High sales:", part1.var())
print("Low sales:", part2.var())
```

**Calculate Standard Deviation**

```python
std = np.sqrt(( part1.var() +  part2.var()) / 2)
print('standard deviation is:', std)
```

**Calculate test statistic**

```python
t = (part1.mean() - part2.mean()) /(std *np.sqrt(2/N))
print('t:', t)
```

**Calculate Degree of Freedom**

```python
df = 2 * N - 2
df
```

**Find the p-value**

```python
p = 1 - stats.t.cdf(abs(t), df=df)  # #p-value after comparison with the t 
print("t = " + str(t))
print("p = " + str(2 * p))
```

```python
null_hypothesis = {
    'H0': "Total price does not increase for Sale",
    'H1': "Total price increases for Large Sale"
}

# ANSWER
def print_sig(p_value, alpha):
    '''
    Inputs:
    p_value: Calculated p_value
    alpha: Confidence level
    '''
    if p_value < alpha:
        print("We reject our null hypothesis.")
        print(null_hypothesis['H1'])
    elif p_value > alpha:
        print("We fail to reject our null hypothesis.")
        print(null_hypothesis['H0'])
    else:
        print("Our test is inconclusive.")
        
# ANSWER
print("t = " + str(t))
print("p = " + str(p*2))
print_sig(p*2, alpha)
```

**Calculate the T-test for the means of two independent samples of scores**

```python
t2, p2 = stats.ttest_ind(part1, part2)
print("t = " + str(t2))
print("p = " + str(p2))
print_sig(p2, alpha)
```
